## WORD2VEC (CBOW)

```python
from gensim.models import Word2Vec

# Contoh data (kalimat sederhana)
sentences = [["saya", "suka", "makan", "apel", "segar"],
             ["apel", "adalah", "buah", "yang", "segar"]]

# Model CBOW dengan Word2Vec
model = Word2Vec(sentences, vector_size=2, window=3, sg=0, min_count=1)

# Melihat hasil vektor dari kata
print("Vektor untuk 'makan':", model.wv["makan"])
print("Vektor untuk 'apel':", model.wv["apel"])

# Melihat kesamaan antara kata
similarity = model.wv.similarity("makan", "apel")
print("Kesamaan antara 'makan' dan 'apel':", similarity)
```

## WORD2VEC (SKIP-GRAM)

```python
from gensim.models import Word2Vec

# Contoh kalimat
sentences = [["saya", "suka", "makan", "apel", "segar"]]

# Membangun model Skip-gram dengan ukuran embedding 2 dan window size 2
model = Word2Vec(sentences, vector_size=2, window=2, sg=1, min_count=1)

# Melihat hasil vektor dari kata-kata
for word in ["saya", "suka", "makan", "apel", "segar"]:
    print(f"Vektor '{word}': {model.wv[word]}")

# Menemukan kata-kata yang paling mirip dengan "suka"
print("\nKata yang mirip dengan 'suka':")
print(model.wv.most_similar("suka"))
```
