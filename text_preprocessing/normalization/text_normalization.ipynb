```python
# contoh teks
teks = 'H-9. Mau ke bank rame bgt ampe antrian tempat duduknya ada diluar. Toko emas meluber ampe ga keliatan mbak2nya yg jualan. Ke swalayan antrian kasir naudzubilah panjang bener. Pada bahagia semua kayaknya. Semoga oh semoga RS sepi, pasien pada sembuh dan pulang utk berbahagia'
```

# Ubah Slangword jadi kata formal

```python
import re

def replace_slang_words(text):
    text = text.lower() # casefolding
    text = re.sub(r'\n', ' ', text) # menghilangkan newline
    text = re.sub(r'\s+', ' ', text) # menghilangkan spasi double
    words = text.split() # memisahkan teks menjadi kata kata
    new_words = [] # untuk menampung kata kata slang yang sudah di rubah 
    for word in words: # proses looping untuk merubah slang word
        if word in slangword: # pengecekan slang word
            new_words.append(slangword[word])
        else:
            new_words.append(word)
    new_text = ' '.join(new_words)
    return new_text
```

```python
teks_formal = replace_slang_words(teks)
print("Teks sebelum menghilangkan slangword : ",teks)
print("Teks sesudah menghilangkan slangword : ", teks_formal)
```

# Steming

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
teks_stem = stemmer.stem(teks_formal)
print("Teks sebelum stemming : ",teks_formal)
print("Teks sesudah stemming : ", teks_stem)
```

# Lemma

### NLP_ID (Kumparan)

```python
from nlp_id.lemmatizer import Lemmatizer 
lemmatizer = Lemmatizer() 

teks_lemma = lemmatizer.lemmatize(teks_formal)

print("Teks sebelum lemma : ",teks_formal)
print("Teks sesudah lemma : ", teks_lemma)
```
